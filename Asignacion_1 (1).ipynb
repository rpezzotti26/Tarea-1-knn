```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score


df = pd.read_csv('/content/cleaned_dataset.csv')
print("¡Archivo detectado y cargado correctamente!")


train_small = df.head(20)
test_small = df.iloc[20:40]


def distancia_euclidiana(v1, v2):
    return np.sqrt(np.sum((np.array(v1) - np.array(v2))**2))


v1 = [1, 106, 70, 28, 135, 34.2, 0.142, 22]
v2 = [2, 102, 86, 36, 120, 45.5, 0.127, 23]
dist = distancia_euclidiana(v1, v2)
print(f"\nDistancia Euclidiana entre v1 y v2: {dist:.4f}")


X = df.drop('Outcome', axis=1)
y = df['Outcome']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, stratify=y, random_state=42)

#
def ejecutar_knn(X_entreno, X_val):
    knn = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
    knn.fit(X_entreno, y_train)
    predicciones = knn.predict(X_val)
    return accuracy_score(y_test, predicciones)


acc_crudo = ejecutar_knn(X_train, X_test)


scaler_norm = MinMaxScaler()
X_train_norm = scaler_norm.fit_transform(X_train)
X_test_norm = scaler_norm.transform(X_test)
acc_norm = ejecutar_knn(X_train_norm, X_test_norm)


scaler_std = StandardScaler()
X_train_std = scaler_std.fit_transform(X_train)
X_test_std = scaler_std.transform(X_test)
acc_std = ejecutar_knn(X_train_std, X_test_std)


print("\n--- TABLA COMPARATIVA DE ACCURACIES ---")
resultados = pd.DataFrame({
    'Metodología': ['KNN sin escalar (80/20)', 'KNN normalizado (80/20)', 'KNN estandarizado (80/20)'],
    'Accuracy (Precisión)': [acc_crudo, acc_norm, acc_std]
})
print(resultados)



print("--- PASO 4: COMPARATIVA REAL VS PREDICHO (10 MUESTRAS) ---")


muestras_prueba = X_test.iloc[:10]
valores_reales = y_test.iloc[:10].values


predicciones_10 = KNeighborsClassifier(n_neighbors=3).fit(X_train, y_train).predict(muestras_prueba)


tabla_paso4 = pd.DataFrame({
    'Muestra': range(1, 11),
    'Valor Real (Resultado)': valores_reales,
    'Valor Predicho': predicciones_10
})

print(tabla_paso4)


!git config --global user.email "robertopezzotti143@gmail.com"
!git config --global user.name "rpezzotti26"

USUARIO = "rpezzotti26"
TOKEN = "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"
REPO = "Tarea-1-knn"


%cd /content/
!rm -rf {REPO}
repo_url = f"https://{TOKEN}@github.com/{USUARIO}/{REPO}.git"
!git clone {repo_url}


!cp "/content/Asignacion-1-IA" "/content/{REPO}/Asignacion-1-IA.ipynb"
!cp "/content/cleaned_dataset.csv" "/content/{REPO}/"


%cd {REPO}
!git add .
!git commit -m "Subida de código y datos corregida"
!git push origin main
```
